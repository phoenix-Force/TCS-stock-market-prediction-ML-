```python
adf=df[df.Symbol=="TCS"]
```

```python
adf.drop(columns=['Symbol'])

sns.heatmap(adf.isnull(),yticklabels=False)
```

```python
adf["diff"]=adf["High"]-adf["Low"]
```

```python
adf.head()
adf.dropna(inplace= True)
```

```python
adf["diffp"] = ((adf["High"]-adf["Low"])/adf["High"])*100
```

```python
adf["close_diff"] = adf["Close"]-adf["Adj Close"]
adf["Close_diff_per"] = ((adf["Close"]-adf["Adj Close"])/adf["Close"])*100

adf.head()
```

```python

adf.head()
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
sns.heatmap(adf.isnull(),yticklabels=False)
```

```python
'''trainx = adf[:["High","Low"."Close","Adj Close","Volume","diff","diffp","close_diff","close_diff_per"]]'''
'''adf.loc[:, 'High':'Close_diff_per']'''

train = adf[:2300]
test = adf[2300:]

x = train.loc[:, 'High':'Close_diff_per']
y = train.loc[:,"Open"]
xx = test.loc[:, 'High':'Close_diff_per']
xy = test.loc[:,"Open"]
xxx=xx.iloc[0]
xy
```

```python
model.fit(x,y)
```

```python
p = model.predict(xx)
p
```

```python
s = model.score(x,y)
```

```python
print("Prediction Score is "+str(s*100))
```
